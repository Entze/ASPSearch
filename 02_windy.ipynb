```python
import clingo
```

```python
inst1 = """

e(2,1,100).
e(3,2,100).
e(1,3,100).
e(1,2,10).
e(2,3,10).
e(3,1,10).

v(U) :- e(U,_,_).
v(V) :- e(_,V,_).
time(0..M) :- maxttime(M).
maxttime(3).

"""
```

```python
inst2 = """

e( 1, 2,1).
e( 1,13,1).
e( 2, 3,1).
e( 2,14,1).
e( 3, 4,1).
e( 3, 5,1).
e( 4, 5,1).
e( 5, 6,1).
e( 5, 7,1).
e( 6, 7,1).
e( 6,14,1).
e( 7, 8,1).
e( 8, 9,1).
e( 8,12,1).
e( 9,10,1).
e( 9,11,1).
e( 9,12,1).
e(10,11,1).
e(12,13,1).
e(12,14,1).
e(13,14,1).

v(U) :- e(U,_,_).
v(V) :- e(_,V,_).

time(0..M) :- maxttime(M).
maxttime(26).

"""
```

```python
gac = """

#program guess.

1 { visitNodeAtTime(V,T) : v(V) } 1 :- time(T).
visited(U,V) :- visitNodeAtTime(U,T), visitNodeAtTime(V,T+1).

#program check.

#external visited(V,U) : v(V),v(U).

bad :- v(U), v(V), visited(U,V), not e(U,V,_).
bad :- v(U), v(V), e(U,V,_), not visited(U,V), not visited(V,U).

% The tour must be closed, i.e. first and last vertex are the same (choose `1`).
bad :- not visitNodeAtTime(1,0).
bad :- not visitNodeAtTime(1,M), maxttime(M).

:- not bad.

"""
```

```python
from gac import Transformer, GACPropagator
from clingo import ast

ctl = clingo.Control(("--models", "0"))

check = []
with ast.ProgramBuilder(ctl) as builder:
    trans = Transformer(builder, check)
    for file in [
                 "#program guess."+inst1,
                 gac,
                 "#program check."+inst1,
                 "#show visitNodeAtTime/2. #show visited/2.", "#show v/1.",
                 ]:
        ast.parse_string(file, trans.add)
ctl.register_propagator(GACPropagator(check))

ctl.ground([("base", [])])
ctl.solve(on_model=print)
```
